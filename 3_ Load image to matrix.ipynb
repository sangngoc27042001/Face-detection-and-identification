```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
X=[]
y=[]
categories=['not_myface','myface']
for category in categories:
    for img in os.listdir(category):
        real_path=os.path.join(category,img)
        img_array = cv2.imread(real_path)
        new_array = cv2.resize(img_array, (150, 150))
        X.append(new_array)
        label=""
        if(category==categories[0]):
            label=0
        else:
            label=1
        y.append([label])
```

```python
X=np.array(X)
X.shape
```

```python
y=np.array(y)
y.shape
```
