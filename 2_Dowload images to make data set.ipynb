```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
import urllib.request #to download image from internet
def download_image_ipg(url, file_path, file_name):
    fullpath=file_path+file_name+".png"
    urllib.request.urlretrieve(url,fullpath)
```

```python
url="https://wallpapercave.com/wp/wp2437393.jpg"
download_image_ipg(url, '','data')
```

```python
img=cv2.imread('data.png')
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python

face=face_cascade.detectMultiScale(gray, scaleFactor=2, minNeighbors=3)
for(x,y,w,h)in face:
        print(x,y,w,h)
        roi_gray=gray[y:y+h, x:x+w]
        img_item="data2"+".png"
        cv2.imwrite(img_item,roi_gray)
        cv2.rectangle(gray, (x, y), (x + w, y + h), (255, 0, 0), 5)
```

```python
i=0
x=[]
big_folder="lfw_funneled"
for file in os.listdir(big_folder):
    path = os.path.join(big_folder,file)
    for img in os.listdir(path):
        path2=os.path.join(path,img)
        img=cv2.imread(path2)
        gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        face=face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
        for(x,y,w,h)in face:
            print(x,y,w,h)
            roi_gray=gray[y:y+h, x:x+w]
            img_item="not_myface/my_image"+str(i)+".png"
            cv2.imwrite(img_item,roi_gray)
            cv2.rectangle(gray, (x, y), (x + w, y + h), (255, 0, 0), 5)
            i=i+1
```
