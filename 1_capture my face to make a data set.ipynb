```python
import numpy as np
import cv2
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cap = cv2.VideoCapture(0)
```

# Capture my face and put these pictures in one folder

```python
i=5165;
while True:
    ret, frame= cap.read()
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    face=face_cascade.detectMultiScale(gray, scaleFactor=2, minNeighbors=5)
    for(x,y,w,h)in face:
        print(x,y,w,h), (255,255,255), 
        roi_gray=gray[y:y+h, x:x+w]
        img_item="myface/my_image"+str(i)+".png"
        cv2.imwrite(img_item,roi_gray)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 5)
        i=i+1
        
    cv2.putText(frame, "sang", (x,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0) )
    cv2.imshow('cak', frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
    
```

```python
cap.release()
cv2.destroyAllWindows()
```
